```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# Addition

```python
# Vectors in R^2
v1 = np.array([2, 3])
v2 = np.array([1, 4])

v3 = v1 + v2
```

```python
# Plotting
plt.plot([0, v1[0]],[0, v1[1]],'b',label='v1')
plt.plot([0, v2[0]]+v1[0],[0, v2[1]]+v1[1],'r',label='v2')
plt.plot([0, v3[0]],[0, v3[1]],'k',label='v1+v2')
plt.legend()
plt.axis('square')
plt.axis((-6, 6, -6, 6 ))
plt.grid()
plt.show()
```

```python
# Vectors in R^3
v1_3 = np.array([2, 3, 5])
v2_3 = np.array([1, 4, 8])

v3_3 = v1_3 + v2_3
```

```python
fig = plt.figure(figsize=plt.figaspect(1))
ax = plt.axes(None, projection='3d')
ax.plot([0, v1_3[0]],[0, v1_3[1]],[0, v1_3[2]],linewidth=3)
ax.plot([0, v2_3[0]],[0, v2_3[1]],[0, v2_3[2]],linewidth=3)
ax.plot([0, v3_3[0]],[0, v3_3[1]],[0, v3_3[2]],linewidth=3)

ax.plot([0, 0],[0, 0],[-4, 4],'k--')
ax.plot([0, 0],[-4, 4],[0, 0],'k--')
ax.plot([-4, 4],[0, 0],[0, 0],'k--')
plt.show()
```

# Subtraction

```python
# Vectors in R^2
v1 = np.array([4, 6])
v2 = np.array([2, 4])

v3 = np.subtract(v1, v2)
```

```python
# Plotting
plt.plot([0, v1[0]],[0, v1[1]],'b',label='v1')
plt.plot([0, v2[0]],[0, v2[1]],'r',label='v2')
plt.plot([0, v3[0]],[0, v3[1]],'k',label='v1-v2')
plt.legend()
plt.axis('square')
plt.axis((-6, 6, -6, 6 ))
plt.grid()
plt.show()
```

```python
# Vectors in R^3
v1_3 = np.array([2, 3, 5])
v2_3 = np.array([1, 4, 8])

v3_3 = v1_3 - v2_3
```

```python
fig = plt.figure(figsize=plt.figaspect(1))
ax = plt.axes(None, projection='3d')
ax.plot([0, v1_3[0]],[0, v1_3[1]],[0, v1_3[2]],linewidth=3)
ax.plot([0, v2_3[0]],[0, v2_3[1]],[0, v2_3[2]],linewidth=3)
ax.plot([0, v3_3[0]],[0, v3_3[1]],[0, v3_3[2]],linewidth=3)

ax.plot([0, 0],[0, 0],[-4, 4],'k--')
ax.plot([0, 0],[-4, 4],[0, 0],'k--')
ax.plot([-4, 4],[0, 0],[0, 0],'k--')
plt.show()
```
